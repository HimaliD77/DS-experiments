```python
from scipy.stats import zscore
```

```python
z_scores = np.abs(df.apply(zscore))
outliers = (z_scores > 3).sum()
```

```python
print("\nOutlier Detection (Z-scores > 3):")
print(outliers)
```

```python
z_scores = df.apply(zscore)

outliers_zscore = (np.abs(z_scores) > 3).sum()
print("\nOutliers Detected Using Z-Score (per column):")
print(outliers_zscore)

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outliers_iqr = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
print("\nOutliers Detected Using IQR (per column):")
print(outliers_iqr)

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient="h", palette="Set2")
plt.title("Boxplot for Outlier Detection")
plt.tight_layout()
plt.show()

```

REMOVE OUTLIERS


```python

z_scores = df.apply(zscore)

data_cleaned_zscore = df[(np.abs(z_scores) <= 3).all(axis=1)]

print("\nDataset After Removing Outliers (Z-Score Method):")
print(data_cleaned_zscore.shape)

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

data_cleaned_iqr = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

print("\nDataset After Removing Outliers (IQR Method):")
print(data_cleaned_iqr.shape)

cleaned_file_path_zscore = "wine_cleaned_zscore.xlsx"
cleaned_file_path_iqr = "wine_cleaned_iqr.xlsx"

data_cleaned_zscore.to_excel(cleaned_file_path_zscore, index=False)
data_cleaned_iqr.to_excel(cleaned_file_path_iqr, index=False)

print(f"\nCleaned dataset (Z-Score) saved as {cleaned_file_path_zscore}")
print(f"Cleaned dataset (IQR) saved as {cleaned_file_path_iqr}")
```
